Для управления ошикбами, которые козникают в проце20се выполнения программ, в Python используют *специальные* **объекты**, называемые **исключениями**. Этими объектами в простейшем случае могут быть блоки `try-except`.

**Задача 1.** Написать программу, которая выводит таблицу значений гиперболы $y=\frac{a}{x}$ на интервале $x \in [xn;xk]$, который включает точку $O$ (0,0).

```python
xn=float(input('xn='))
xk=float(input('xk='))
a=float(input('a='))
n=int(input('Введите количество интервалов'))
h=(xk-xn)/n
for i in range(0,n+1):
    x=xn+i*h
    y=a/x
    print("x=",x,"y=",y)
```

Программа не будет корректно работать, если среди значений *x* будет число **0.0**.

```python
Чтобы избавиться от этой проблемы, давайте, доавим обрабработку исключения.
```

```python
xn=float(input('xn='))
xk=float(input('xk='))
a=float(input('a='))
n=int(input('Введите количество интервалов'))
h=(xk-xn)/n
for i in range(0,n+1):
    x=xn+i*h
    try:
        y=a/x
    except ZeroDivisionError:
        print("Деление на ноль")
    else:
       print("x=",x,"y=",y)
```

Теперь решим более сложную задачу, построить график гиперболы $$y=\frac{a}{x}$$ на интервале $x \in [xn;xk]$, который включает точку $O$ (0,0).

```python
import matplotlib.pyplot as plt
import sys
def printf(format, *args):
    sys.stdout.write(format % args)
xn=-1
xk=-1
while(xn*xk>0):
    xn=float(input('xn='))
    xk=float(input('xk='))
if xn>xk:
    z=xn
    xn=xk
    xk=z
a=float(input('a='))
n=int(input('Введите количество интервалов'))
h=(xk-xn)/n
X=[];
Y=[];
X1=[];
Y1=[];
fl=0;
print("xn=",xn,"xk=",xk)
for i in range(0,n+1):
    x=xn+i*h
    try:
        y=a/x
    except ZeroDivisionError:
        print("Деление на ноль")
        fl=1;
    else:
        if (fl==0) and (x*xn<0):
            fl=1;
        printf("x=%8.2f\ty=%8.2f\n",x,y)
        if (fl):
            X.append(x)
            Y.append(y)
        else:
            X1.append(x)
            Y1.append(y)   
#print(X)
#print(Y)
plt.plot(X,Y,'b-',X1,Y1,'k-')
plt.grid(True)
plt.show()
```

Вернёмся к примеру деления и посмотрим возможные варианты обработки исключений

```python
try:
    x=1/int(input('x='))
#Проверка первого исключения, деления на 0.
except ZeroDivisionError:
    print("Ошибка деления на ноль.")
#Проверка второго исключения, преобразования введенного значения в вещественное число или строку
except ValueError:
    print("Невозможно преобразовать строку в число.")
#Неизвестное программисту второе исключение
except Exception:
    print("Неизвестная ошибка.")
# обработка случая корректной обработки, нет исключения.
else:
    print("Операция выполнена успешно.")
    print("x=",x)
# Код выполяемый в любом случае)))
finally:
    print("Программа завершена.")
```
