## Пример обычной таблицы

| Выравнивание влево | Выравнивание по центру | Выравнивание по правой границы |
|:------------- |:---------------:| -------------:|
| Строка 1         | **Полужирный**        | Ячейка 3 |
| Строка 2         | *Курсив*        | Ячейка 6 |
| Строка 4         | ~~Зачёркнуты1~~      | Ячейка 9 |
| строка 4         | [Ссылка](dot.com) | Ячейка 12 |
